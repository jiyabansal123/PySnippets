```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow  # For displaying images in Colab

# Step 1: Upload the image
from google.colab import files
uploaded = files.upload()  # Choose your image file here

# Get uploaded image filename
filename = list(uploaded.keys())[0]

# Step 2: Load the image using OpenCV
image_bgr = cv2.imread(filename)
print("Original Image Shape:", image_bgr.shape)

# Step 3: Convert BGR to RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Display the RGB image
plt.figure(figsize=(4,4))
plt.title("RGB Image")
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Step 4: Resize image to 128x128
image_resized = cv2.resize(image_rgb, (128, 128))
print("Resized Image Shape:", image_resized.shape)

# Step 5: Convert to Grayscale
image_gray = cv2.cvtColor(image_resized, cv2.COLOR_RGB2GRAY)

plt.title("Grayscale Image")
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.show()

# Step 6: Convert grayscale to binary using thresholding
_, image_binary = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)

plt.title("Binary Image (Thresholding)")
plt.imshow(image_binary, cmap='gray')
plt.axis('off')
plt.show()

# Step 7: Normalize the image pixel values (0 to 1)
image_normalized = image_resized / 255.0

plt.title("Normalized RGB Image")
plt.imshow(image_normalized)
plt.axis('off')
plt.show()

# Step 8: Apply basic augmentations
# Horizontal Flip
image_flipped = cv2.flip(image_resized, 1)
# Rotate 90 degrees clockwise
image_rotated = cv2.rotate(image_resized, cv2.ROTATE_90_CLOCKWISE)

# Display flipped and rotated images
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Flipped Image")
plt.imshow(image_flipped)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Rotated Image")
plt.imshow(image_rotated)
plt.axis('off')
plt.show()

# Step 9: Apply Gaussian and Median Blurs
image_gaussian = cv2.GaussianBlur(image_resized, (5, 5), 0)
image_median = cv2.medianBlur(image_resized, 5)

# Display filtered images
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Gaussian Blur")
plt.imshow(image_gaussian)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Median Blur")
plt.imshow(image_median)
plt.axis('off')
plt.show()

# Step 10: Flatten the grayscale image
flattened = image_gray.flatten()
print("Flattened Image Shape:", flattened.shape)
```
